```python
class Patient:
  def __init__(self,id_no,name,age,blood):
    self.id=id_no
    self.name=name
    self.age=age
    self.blood_group=blood

class DoublyNode:
    def __init__(self,el,n,p):
        self.elem=el
        self.next=n
        self.prev=p
class WRM:
  def __init__(self):
    self.head=DoublyNode(None, None, None)
    self.head.next=self.head
    self.head.prev=self.head

  def RegisterPatient(self, patient):
    new=DoublyNode(patient, None, None)
    self.head.prev.next=new
    new.prev=self.head.prev
    new.next=self.head
    self.head.prev=new
    print("Successfully registered patient information")

  def ServePatient(self):
    if self.head.next==self.head:
      print("No patient to serve")
      return
    p=self.head.next
    print(p.elem.name,"is being served")
    p.next.prev=self.head
    self.head.next=p.next

  def ShowAllPatient(self):
    if self.head.next==self.head:
      return
    head=self.head.next
    while True:
      print(head.elem.name,f"Age-{head.elem.age}",f"ID-{head.elem.id}",f"Blood-Group-{head.elem.blood_group}")
      head=head.next
      if head==self.head:
        break

  def CancelAll(self):
    self.__init__()

  def CanDoctorGoHome(self):
    if self.head.next==self.head:
      return True
      return False

  def ReverseTheLine(self):
    if self.head==self.head.next:
      print('Sorry! No Patient found.')
    else:
      new=self.head
      while True:
        temp1=new.prev
        temp2=new.next
        new.next=temp1
        new.prev=temp2
        new=new.prev
        if new==self.head:
          break

    print("Reverse line Successful")

y = WRM()
while True:
    print("Enter command:")
    print("  1. Register")
    print("  2. Serve")
    print("  3. ShowAll")
    print("  4. CancelAll")
    print("  5. Can go home?")
    print("  6. Reverse the Line")
    print("  7. Quit")
    x =input().strip()
    if x=="1":
        patient = Patient(input("Enter ID: "),input("Enter name: "),input("Enter age: "),input("Enter blood group: "))
        y.RegisterPatient(patient)
    elif x=="2":
        y.ServePatient()
    elif x=="3":
        y.ShowAllPatient()
    elif x=="4":
        y.CancelAll()
    elif x=="5":
        if y.CanDoctorGoHome():
            print("YES")
        else:
            print("NO")
    elif x=="6":
        y.ReverseTheLine()
    elif x=="7":
        break
    else:
        print("Wrong command!!")
```
